# 1) MNIST GAN

# MNIST GAN — Generating Handwritten Digits

In this notebook, we build a **Generative Adversarial Network (GAN)** to generate new handwritten digit images similar to those in the **MNIST dataset**.

### What is MNIST?
MNIST is a classic dataset of handwritten digits:
- 70,000 grayscale images
- Image size: 28 × 28 pixels
- Digits: 0 → 9

It is commonly used as the "hello world" dataset of computer vision.

### What is a GAN?
A GAN consists of two neural networks competing with each other:

**1. Generator (G):**  
- Takes random noise as input.  
- Tries to generate fake images that look real.

**2. Discriminator (D):**  
- Receives real images from MNIST and fake images from the Generator.  
- Tries to distinguish real vs fake.

### Training Idea
- The Generator learns to fool the Discriminator.
- The Discriminator learns to detect fake images.
- Over time, the Generator produces realistic handwritten digits.

### Goal
By the end of training, the Generator can create new digit images that resemble real MNIST samples — even though they never existed in the dataset.

```python
# ==================================================
# 1) Imports
# ==================================================
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# ==================================================
# 2) Hyperparameters
# ==================================================
latent_dim = 100
batch_size = 64
epochs = 10
learning_rate = 0.0002

# ==================================================
# 3) Dataset Loading
# ==================================================
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))  # Scale pixels to [-1,1]
])

train_dataset = datasets.MNIST(
    root="./data",
    train=True,
    transform=transform,
    download=True
)

train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True
)

# ==================================================
# 4) Generator Model
# ==================================================
class Generator(nn.Module):
    """
    Generator: Converts random noise (z) into fake MNIST images.
    Input:  z -> [batch_size, latent_dim]
    Output: fake image -> [batch_size, 1, 28, 28]
    """
    def __init__(self, latent_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(latent_dim, 256), #  بعمل توسيع من 100 ل 256
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.ReLU(True),
            nn.Linear(512, 784),
            nn.Tanh()
        )

    def forward(self, z):
        img = self.net(z)
        return img.view(z.size(0), 1, 28, 28)

# ==================================================
# 5) Discriminator Model
# ==================================================
class Discriminator(nn.Module):
    """
    Discriminator: Classifies images as real or fake.
    Input:  image -> [batch_size, 1, 28, 28]
    Output: probability -> [batch_size, 1]
    """
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(784, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, img):
        img_flat = img.view(img.size(0), -1)
        return self.net(img_flat)

# ==================================================
# 6) Initialize Models, Loss, Optimizers
# ==================================================
G = Generator(latent_dim)
D = Discriminator()

loss_fn = nn.BCELoss()

optimizer_G = optim.Adam(G.parameters(), lr=learning_rate)
optimizer_D = optim.Adam(D.parameters(), lr=learning_rate)

# ==================================================
# 7) Utility: Function to Visualize Generated Images
# ==================================================
def show_generated_images(generator, num_images=16):
    generator.eval()
    z = torch.randn(num_images, latent_dim)
    fake_imgs = generator(z).detach()

    grid = fake_imgs.view(num_images, 28, 28)

    plt.figure(figsize=(4,4))
    for i in range(num_images):
        plt.subplot(4,4,i+1)
        plt.imshow(grid[i], cmap='gray')
        plt.axis('off')
    plt.suptitle("Generated Images")
    plt.show()
    generator.train()

# ==================================================
# 8) Training Loop
# ==================================================
for epoch in range(epochs):
    for batch_idx, (real_imgs, _) in enumerate(train_loader):

        batch_size_now = real_imgs.size(0)

        # Labels
        real_labels = torch.ones(batch_size_now, 1)
        fake_labels = torch.zeros(batch_size_now, 1)

        # ---------------------
        # Train Discriminator
        # ---------------------
        optimizer_D.zero_grad()

        # Real images
        real_output = D(real_imgs)  # real = 1
        loss_real = loss_fn(real_output, real_labels)

        # Fake images
        z = torch.randn(batch_size_now, latent_dim)
        fake_imgs = G(z) # Gen
        fake_output = D(fake_imgs.detach()) # D
        loss_fake = loss_fn(fake_output, fake_labels)

        # Total Discriminator Loss
        d_loss = loss_real + loss_fake
        d_loss.backward()
        optimizer_D.step() # update D only Backprog

        # ---------------------
        # Train Generator
        # ---------------------
        optimizer_G.zero_grad()

        z = torch.randn(batch_size_now, latent_dim) # 10*10
        fake_imgs = G(z) # G(10*10) >> output 28*28 fake img  20 img
        output = D(fake_imgs) # P(Real) = 0.8

        # Generator wants discriminator to believe fakes are real
        g_loss = loss_fn(output, real_labels) # D = 0.2 > fake >  بدي اعاقب ال G
        g_loss.backward() # evaluator
        optimizer_G.step() # update G only Backprog

        # Print progress occasionally
        if batch_idx % 200 == 0:
            print(f"Epoch [{epoch+1}/{epochs}] Batch [{batch_idx}/{len(train_loader)}] | D Loss: {d_loss.item():.4f} | G Loss: {g_loss.item():.4f}")

    # Show generated samples after each epoch
    show_generated_images(G)

print("Training Complete!")
```

```python
# ==========================================
# Save the trained Generator model
# ==========================================

# After training is finished, we save only the Generator.
# The Discriminator is no longer needed.

torch.save(G.state_dict(), "generator_mnist.pth")

print("Generator model saved successfully!")
```

```python
# ==========================================
# Load the Generator for future use
# ==========================================

# Recreate the Generator architecture
loaded_G = Generator(latent_dim)

# Load saved weights
loaded_G.load_state_dict(torch.load("generator_mnist.pth"))

# Set to evaluation mode (no training, no dropout, no batchnorm updates)
loaded_G.eval()

print("Generator model loaded and ready for use!")
```

```python
# ==========================================
# Generate new images from random noise
# ==========================================

num_samples = 16

# Create new random noise vectors
z = torch.randn(num_samples, latent_dim)

# Generate fake images
with torch.no_grad():  # No gradients needed
    generated_images = loaded_G(z)

print("Generated images shape:", generated_images.shape)
# Output: [16, 1, 28, 28]
```

```python
# ==========================================
# Display generated images
# ==========================================

import matplotlib.pyplot as plt

generated_images = generated_images.squeeze().numpy()

plt.figure(figsize=(4,4))
for i in range(num_samples):
    plt.subplot(4,4,i+1)
    plt.imshow(generated_images[i], cmap="gray")
    plt.axis("off")

plt.suptitle("New Images Generated by GAN")
plt.show()
```

#2) DCGAN on MNIST

## Key Differences Between Basic GAN and DCGAN on MNIST

Although both models follow the same adversarial training idea, the **architecture design** is very different.

---

### 1) Network Structure

**Basic GAN:**
- Uses Fully Connected (Linear) layers only.
- Flattens images into 1D vectors.
- Does not explicitly preserve spatial structure.

**DCGAN:**
- Uses Convolutional and Transposed Convolutional layers.
- Keeps images in 2D format (Height × Width × Channels).
- Learns spatial features like edges and strokes.

**Impact:**  
DCGAN generates sharper and more realistic images.

---

### 2) Image Representation

**Basic GAN:**
- Treats a 28×28 image as a vector of 784 numbers.
- Ignores pixel neighborhood relationships.

**DCGAN:**
- Treats the image as a 2D grid.
- Nearby pixels influence each other through convolution filters.

**Impact:**  
Better understanding of image structure.

---

### 3) Upsampling Method in Generator

**Basic GAN:**
- Uses Linear layers to expand noise vector to image size.

**DCGAN:**
- Uses ConvTranspose2D layers to gradually upscale:
  - 7×7 → 14×14 → 28×28

**Impact:**  
Smoother and more stable image generation.

---

### 4) Feature Extraction in Discriminator

**Basic GAN:**
- Uses Linear layers to classify images.
- Needs to learn everything from flattened pixels.

**DCGAN:**
- Uses Conv2D layers to extract visual features automatically.

**Impact:**  
Stronger discriminator → better generator learning.

---

### 5) Training Stability

**Basic GAN:**
- More prone to instability and mode collapse.

**DCGAN:**
- Uses Batch Normalization and convolutional design.
- More stable and faster convergence.

---

### 6) Output Quality

**Basic GAN:**
- Generated digits may look blurry or inconsistent.

**DCGAN:**
- Generated digits have clearer strokes and shapes.

---

### Summary Table

| Aspect | Basic GAN | DCGAN |
|--------|-----------|--------|
| Layers | Fully Connected | Convolutional |
| Image Structure | Flattened | Preserved |
| Upsampling | Linear Expansion | ConvTranspose2D |
| Feature Learning | Weak | Strong |
| Image Quality | Lower | Higher |
| Training Stability | Harder | Easier |

---

**Conclusion:**  
DCGAN is a specialized GAN architecture designed specifically for image generation, which is why it significantly outperforms a basic GAN on MNIST.

```python
# ==================================================
# 1) Imports
# ==================================================
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# ==================================================
# 2) Device Setup (GPU if available)
# ==================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ==================================================
# 3) Hyperparameters
# ==================================================
latent_dim = 100
batch_size = 128
epochs = 20
learning_rate = 0.0002

# ==================================================
# 4) Dataset Loading
# ==================================================
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_dataset = datasets.MNIST(
    root="./data",
    train=True,
    transform=transform,
    download=True
)

train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True
)

# ==================================================
# 5) DCGAN Generator
# ==================================================
class Generator(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(latent_dim, 128 * 7 * 7),
            nn.ReLU(True),
            nn.Unflatten(1, (128, 7, 7)),
            # conv2d
            nn.ConvTranspose2d(128, 64, 4, 2, 1),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 1, 4, 2, 1),
            nn.Tanh()
        )

    def forward(self, z):
        return self.net(z)

# ==================================================
# 6) DCGAN Discriminator
# ==================================================
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(1, 64, 4, 2, 1),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, 128, 4, 2, 1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Flatten(),
            nn.Linear(128 * 7 * 7, 1),
            nn.Sigmoid()
        )

    def forward(self, img):
        return self.net(img)

# ==================================================
# 7) Initialize Models on GPU
# ==================================================
G = Generator(latent_dim).to(device)
D = Discriminator().to(device)

loss_fn = nn.BCELoss()

optimizer_G = optim.Adam(G.parameters(), lr=learning_rate, betas=(0.5, 0.999))
optimizer_D = optim.Adam(D.parameters(), lr=learning_rate, betas=(0.5, 0.999))

# ==================================================
# 8) Utility: Show Generated Images
# ==================================================
def show_generated_images(generator, num_images=16):
    generator.eval()
    z = torch.randn(num_images, latent_dim).to(device)
    with torch.no_grad():
        fake_imgs = generator(z)

    fake_imgs = fake_imgs.squeeze().cpu().numpy()

    plt.figure(figsize=(4,4))
    for i in range(num_images):
        plt.subplot(4,4,i+1)
        plt.imshow(fake_imgs[i], cmap='gray')
        plt.axis('off')
    plt.suptitle("DCGAN Generated Images")
    plt.show()
    generator.train()

# ==================================================
# 9) Training Loop (GPU Enabled)
# ==================================================
for epoch in range(epochs):
    for batch_idx, (real_imgs, _) in enumerate(train_loader):

        real_imgs = real_imgs.to(device)
        batch_size_now = real_imgs.size(0)

        real_labels = torch.ones(batch_size_now, 1).to(device)
        fake_labels = torch.zeros(batch_size_now, 1).to(device)

        # ---------------------
        # Train Discriminator
        # ---------------------
        optimizer_D.zero_grad()

        real_output = D(real_imgs)
        loss_real = loss_fn(real_output, real_labels)

        z = torch.randn(batch_size_now, latent_dim).to(device)
        fake_imgs = G(z)
        fake_output = D(fake_imgs.detach())
        loss_fake = loss_fn(fake_output, fake_labels)

        d_loss = loss_real + loss_fake
        d_loss.backward()
        optimizer_D.step()

        # ---------------------
        # Train Generator
        # ---------------------
        optimizer_G.zero_grad()

        z = torch.randn(batch_size_now, latent_dim).to(device)
        fake_imgs = G(z)
        output = D(fake_imgs)

        g_loss = loss_fn(output, real_labels)
        g_loss.backward()
        optimizer_G.step()

        if batch_idx % 200 == 0:
            print(f"Epoch [{epoch+1}/{epochs}] Batch [{batch_idx}/{len(train_loader)}] | D Loss: {d_loss.item():.4f} | G Loss: {g_loss.item():.4f}")

    show_generated_images(G)

print("DCGAN Training Complete!")
```

```python
# ==========================================
# Save trained DCGAN Generator
# ==========================================
# G.state_dict() returns a dictionary containing all learned parameters (weights and biases) of the Generator.
# This represents the trained "memory" of the model that we save to disk.
torch.save(G.state_dict(), "dcgan_generator_mnist.pth")

print("Generator saved successfully!")
```

```python
# ==========================================
# Load DCGAN Generator for inference
# ==========================================

loaded_G = Generator(latent_dim)
loaded_G.load_state_dict(torch.load("dcgan_generator_mnist.pth"))

loaded_G.eval()   # Switch to inference mode

print("Generator loaded and ready!")
```

```python
# ==========================================
# Generate new MNIST-like images
# ==========================================

num_samples = 16

z = torch.randn(num_samples, latent_dim)   # New random noise

with torch.no_grad():                       # No gradients needed
    new_images = loaded_G(z)

print("Generated images shape:", new_images.shape)
# Expected: [16, 1, 28, 28]
```

### ConvTranspose2D Output Size Formula

For each spatial dimension:

Output = (Input - 1) * Stride - 2 * Padding + Kernel_Size + Output_Padding

---

**Example from our model:**

Input = 7  
Stride = 2  
Padding = 1  
Kernel Size = 4  
Output Padding = 0  

Calculation:

(7 - 1) * 2 - 2 * 1 + 4  
= 6 * 2 - 2 + 4  
= 12 - 2 + 4  
= 14

So:

7 x 7  →  14 x 14
