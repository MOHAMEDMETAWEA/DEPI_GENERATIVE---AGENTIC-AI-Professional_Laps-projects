```python
import requests
import pandas as pd

url = "https://api.nasa.gov/insight_weather/?api_key=DEMO_KEY&feedtype=json&ver=1.0"

r = requests.get(url)
data = r.json()

rows = []

for sol in data["sol_keys"]:
    sol_data = data[sol]
    
    row = {
        "Sol": sol,
        "Avg Temp": sol_data["AT"]["av"],
        "Min Temp": sol_data["AT"]["mn"],
        "Max Temp": sol_data["AT"]["mx"],
        "Wind Speed": sol_data["HWS"]["av"],
        "Pressure": sol_data["PRE"]["av"],
        "Season": sol_data["Season"]
    }
    
    rows.append(row)

df = pd.DataFrame(rows)

df
```

```python
import requests

url='https://api.openweathermap.org/data/2.5/weather'

Parameters={'lat': 31.333 ,
            'lon': 29.444,
            'appid': '9b950a9c77090cb6ca164b660e698619',
            'units': 'metric'}
response=requests.get(url, params=Parameters)

if response.status_code==200:
    data=response.json()
    print(data)
else:
    print(response.status_code)
```
