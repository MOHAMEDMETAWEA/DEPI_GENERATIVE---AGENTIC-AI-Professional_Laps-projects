1. Calculate Average Student Grades
In this exercise, you will create a program that calculates the average grade of students. 
The user inputs the number of students, then enters each student's grade. Use a list to store grades and compute the average.

Hints: Use a for loop to collect grades. Store them in a list. Use sum() and len() to calculate the average.

```python
def calculate_average():
    n = int(input("Number of students: "))
    grades = []
    for i in range(n):
        grade = float(input(f"Grade for student {i+1}: "))
        grades.append(grade)
    avg = sum(grades) / len(grades)
    print(f"Average: {avg:.2f}")

calculate_average()
```

4. Count Occurrences of a Character in a String
Write a function that counts how many times a specific character appears in a string, case-insensitively.

Hints: Convert both to lower case. Use a for loop and a counter.

```python
def count_char(text, char):
    count = 0
    char = char.lower()
    for c in text.lower():
        if c == char:
            count += 1
    return count

print(count_char("Hello World", 'l'))  # 3
```

9. Rock-Paper-Scissors Game
Implement a rock-paper-scissors game against the computer for 3 rounds.

Hints: Use random.choice. For loop for rounds. If-elif for win conditions.

```python
import random

def play_rps():
    choices = ["rock", "paper", "scissors"]
    score = {"player": 0, "computer": 0}
    
    for _ in range(3):
        player = input("rock/paper/scissors: ").strip().lower()
        computer = random.choice(choices)
        print(f"Computer: {computer}")
        
        if player == computer:
            print("Tie!")
        elif (player == "rock" and computer == "scissors") or \
             (player == "paper" and computer == "rock") or \
             (player == "scissors" and computer == "paper"):
            score["player"] += 1
            print("You win!")
        else:
            score["computer"] += 1
            print("You lose!")
    
    print(f"Final score: You {score['player']} - Computer {score['computer']}")

play_rps()
```

11. Employee Points System
Create a system to input employee names and daily points for 5 days, then calculate total and average.

13. Find Duplicates in a List
Write a function to find and return duplicate items in a list.

Hints: Use two sets: seen and duplicates. For loop to check.

```python
def find_duplicates(lst):
    seen = set()
    duplicates = set()
    for item in lst:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)
    return list(duplicates)

print(find_duplicates([1,2,3,2,4,5,3,6,1]))  # [1, 2, 3]
```

14. Simulate Dice Rolls and Count Frequencies
Simulate rolling two dice 10000 times and count frequencies of sums from 2 to 12.

Hints: Use dict for counts. For loop for simulations. random.randint(1,6).

```python
import random

counts = {i: 0 for i in range(2,13)}

for _ in range(10000):
    dice = random.randint(1,6) + random.randint(1,6)
    counts[dice] += 1

for total, count in sorted(counts.items()):
    print(f"{total}: {count / 100:.1f}%")
```

15. Rectangle Class - Area and Perimeter
Create a Rectangle class with methods to calculate area and perimeter.

Hints: Use __init__ for width and height. Simple math in methods.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

r = Rectangle(5, 3)
print(f"Area: {r.area()} | Perimeter: {r.perimeter()}")
```

16. Library System (OOP)
Implement Book and Library classes for adding books, borrowing, and returning.

Hints: Book has available flag. Library has list of books. Methods for borrow/return.

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
        self.available = True
    
    def borrow(self):
        if self.available:
            self.available = False
            return True
        return False
    
    def return_book(self):
        self.available = True

class Library:
    def __init__(self):
        self.books = []
    
    def add_book(self, book):
        self.books.append(book)
    
    def find_book(self, title):
        for b in self.books:
            if b.title.lower() == title.lower():
                return b
        return None

lib = Library()
lib.add_book(Book("1984", "George Orwell", 1949))
lib.add_book(Book("The Mother", "Maxim Gorky", 1906))

book = lib.find_book("1984")
if book and book.borrow():
    print("Borrowed successfully")
```

17. Bank System - Savings Account with Interest (Inheritance)
Create BankAccount base class and SavingsAccount subclass with monthly interest.

Hints: Use super() in subclass. Private balance. Simple interest calculation.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited {amount} → Balance: {self._balance}")
    
    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew {amount} → Balance: {self._balance}")
        else:
            print("Insufficient balance")

class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0, rate=0.05):
        super().__init__(owner, balance)
        self.rate = rate
    
    def add_monthly_interest(self):
        interest = self._balance * self.rate / 12
        self._balance += interest
        print(f"Monthly interest: {interest:.2f} → Balance: {self._balance:.2f}")

acc = SavingsAccount("Lobna", 10000)
acc.deposit(5000)
acc.add_monthly_interest()
```

18. Employee System - Salary with Bonus (Polymorphism)
Base Employee class with calculate_salary, subclasses Manager and Developer overriding it.

Hints: Override method in subclasses. List of employees to demonstrate polymorphism.

```python
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary
    
    def calculate_salary(self):
        return self.base_salary

class Manager(Employee):
    def calculate_salary(self):
        return self.base_salary + 3000  # bonus

class Developer(Employee):
    def __init__(self, name, base_salary, projects):
        super().__init__(name, base_salary)
        self.projects = projects
    
    def calculate_salary(self):
        return self.base_salary + (self.projects * 500)

emps = [
    Manager("Ahmed", 15000),
    Developer("Sarah", 12000, 4),
    Employee("Mohamed", 8000)
]

for e in emps:
    print(f"{e.name}: Salary {e.calculate_salary()}")
```

19. Shopping System with Discounts (OOP + List)
Product class and Cart class to add items and calculate total with category discounts.

Hints: List of tuples in cart. Conditional discount in total method.

```python
class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

class Cart:
    def __init__(self):
        self.items = []
    
    def add(self, product, quantity=1):
        self.items.append((product, quantity))
    
    def total(self):
        total = 0
        for prod, qty in self.items:
            price = prod.price
            if prod.category == "electronics":
                price *= 0.9  # 10% discount
            total += price * qty
        return total

cart = Cart()
cart.add(Product("Laptop", 1000, "electronics"), 1)
cart.add(Product("Book", 20, "books"), 3)
print(f"Total after discount: {cart.total():.2f}")
```

20. Tic Tac Toe Game (OOP)
Implement a TicTacToe class for a simple text-based game.

Hints: List for board. Methods for print, move, check winner. While loop for game.

```python
class TicTacToe:
    def __init__(self):
        self.board = [" " for _ in range(9)]
        self.current_player = "X"
    
    def print_board(self):
        for i in range(0,9,3):
            print("|".join(self.board[i:i+3]))
            if i < 6: print("-+-+- ")
    
    def make_move(self, pos):
        if self.board[pos] == " ":
            self.board[pos] = self.current_player
            self.current_player = "O" if self.current_player == "X" else "X"
            return True
        return False
    
    def check_winner(self):
        wins = [(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
        for a,b,c in wins:
            if self.board[a] == self.board[b] == self.board[c] != " ":
                return self.board[a]
        if " " not in self.board:
            return "Tie"
        return None

game = TicTacToe()
while True:
    game.print_board()
    pos = int(input(f"Player {game.current_player} (0-8): "))
    if 0 <= pos <= 8 and game.make_move(pos):
        winner = game.check_winner()
        if winner:
            game.print_board()
            print("Winner" if winner != "Tie" else "Tie", winner if winner != "Tie" else "")
            break
```
