```python
def calculate_average():
    n=int(input('Number of students'))
    grades=[]

    for i in range(n):
        grade=float(input('enter the grade'))
        grades.append(grade)

    if n ==0:
        print('no students ')

    else:
        print(f'the average of grades = {sum(grades)/ n}')


calculate_average()
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mult(a,b):
    return a*b
def div(a,b):
    if b==0:
        print('invalid divsion')
    else:
        return a/b
    

while True:
    print('1- add / 2- sub  / 3- mult  4- div   0- exit')
    choice= int(input('enter your choice'))


    if choice == 0:
        break
 
    a=float(input('enter the num'))
    b=float(input('enter the num'))

    if choice == 1:
        print(add(a,b))

    elif choice ==2:
        print(sub(a,b))

    elif choice ==3:
        print(mult(a,b))

    elif choice ==4:
        print(div(a,b))
```

```python
def count_char(text, char):
    char= char.lower()
    count=0
    for x in text.lower():
        if x == char:
            count +=1
    return count

count_char('HelLo in python world','l' )
```
