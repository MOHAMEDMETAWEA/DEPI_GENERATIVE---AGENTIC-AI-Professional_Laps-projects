## 1. Lists

**Definition and Properties:**
- A list is an ordered, mutable (changeable) collection of items.
- Can contain duplicates and mixed data types (e.g., integers, strings).
- Indexed (access by position, starting from 0).
- Created with square brackets: `my_list = [1, 'apple', 3.14]`.
- Useful for sequences where order matters and items may change.

**Common Methods:**
- `append(item)`: Add item to the end.
- `insert(index, item)`: Add item at specific position.
- `remove(item)`: Remove first occurrence of item.
- `pop(index)`: Remove and return item at index (default last).
- `sort()`: Sort in place (ascending).
- `reverse()`: Reverse in place.
- `index(item)`: Find position of item.
- `count(item)`: Count occurrences.
- `len(list)`: Get length.
- Slicing: `list[start:end:step]` to get sublist.

## 2. Sets

**Definition and Properties:**
- A set is an unordered, mutable collection of unique items (no duplicates).
- Cannot contain mutable elements like lists, but can have immutable ones like tuples.
- No indexing (cannot access by position).
- Created with curly braces: `my_set = {1, 2, 3}` or `set([1,2,3])`.
- Ideal for membership testing, removing duplicates, and mathematical operations (union, intersection).

**Common Methods:**
- `add(item)`: Add a new item.
- `remove(item)`: Remove item (error if not found).
- `discard(item)`: Remove item (no error if not found).
- `pop()`: Remove and return arbitrary item.
- `clear()`: Empty the set.
- `union(other_set)` or `|`: Combine sets.
- `intersection(other_set)` or `&`: Common items.
- `difference(other_set)` or `-`: Items in first but not second.
- `len(set)`: Get number of items.
- `in` operator: Check membership fast.

## 3. Tuples

**Definition and Properties:**
- A tuple is an ordered, immutable (unchangeable) collection of items.
- Can contain duplicates and mixed types.
- Indexed like lists.
- Created with parentheses: `my_tuple = (1, 'apple', 3.14)`.
- Useful for fixed data, multiple return values from functions, or as dict keys (since immutable).

**Common Methods:**
- `index(item)`: Find position of item.
- `count(item)`: Count occurrences.
- `len(tuple)`: Get length.
- Slicing: `tuple[start:end:step]`.
- No methods to modify (e.g., no append/remove) because immutable.

## 4. Dictionaries

**Definition and Properties:**
- A dictionary is an unordered, mutable collection of key-value pairs.
- Keys must be unique and immutable (e.g., strings, numbers, tuples).
- Values can be any type, including lists or other dicts.
- Created with curly braces: `my_dict = {'key1': 'value1', 'key2': 42}`.
- Useful for mapping (e.g., lookups by key), JSON-like data.

**Common Methods:**
- `get(key, default)`: Get value for key (or default if not found).
- `keys()`: List of keys.
- `values()`: List of values.
- `items()`: List of (key, value) tuples.
- `update(other_dict)`: Merge another dict.
- `pop(key)`: Remove and return value for key.
- `clear()`: Empty the dict.
- `len(dict)`: Number of keys.
- Access: `dict[key]` (error if not found) or `dict.get(key)`.

## Integrated Applications

Now let's combine these data structures with **functions**, **loops**, and **conditions** in practical examples.

**Application 1: Student Grade Tracker**
- Use a dictionary to store students (key: name, value: list of grades)
- Function to add grade
- Loop to calculate average if >3 grades, else 'Not enough data'
- Use set to check unique subjects if needed

```python
# Application 1 Code
students = {}

def add_grade(name, grade):
    if name not in students:
        students[name] = []
    students[name].append(grade)

add_grade('Alice', 85)
add_grade('Alice', 90)
add_grade('Bob', 75)
add_grade('Bob', 80)
add_grade('Bob', 95)

for name, grades in students.items():
    if len(grades) > 2:
        avg = sum(grades) / len(grades)
        print(f"{name}'s average: {avg:.2f}")
    else:
        print(f"{name}: Not enough data")

# Unique grades set
all_grades = set(grade for grades in students.values() for grade in grades)
print("Unique grades:", all_grades)
```

**Application 2: Inventory Management**
- Dictionary for inventory (key: item, value: tuple (quantity, price))
- Function to update stock with loop for multiple items
- Condition to check low stock (<5)
- List to log changes

```python
# Application 2 Code
inventory = {
    'apple': (10, 1.5),
    'banana': (5, 0.5),
    'cherry': (20, 2.0)
}

def update_stock(item, change):
    if item in inventory:
        qty, price = inventory[item]
        new_qty = qty + change
        if new_qty >= 0:
            inventory[item] = (new_qty, price)
            return True
    return False

# Loop to update multiple
updates = [('apple', -2), ('banana', 3), ('cherry', -5)]
log = []
for item, change in updates:
    if update_stock(item, change):
        log.append(f"Updated {item} by {change}")
    else:
        log.append(f"Failed to update {item}")

print(log)

# Check low stock
low_stock = {k: v for k, v in inventory.items() if v[0] < 5}
print("Low stock:", low_stock)
```
