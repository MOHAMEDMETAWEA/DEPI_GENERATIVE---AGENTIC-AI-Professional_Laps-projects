## 1. Introduction to OOP and Why It’s Used

Object-Oriented Programming is a programming paradigm—a way of thinking about and structuring your code. Instead of writing long scripts of procedural instructions, OOP allows us to model real-world things (or "objects") in our code. These objects have both **data** (attributes or properties) and **behaviors** (methods or functions).

### Real-Life Analogy: A Car

Think about a car. A car is an object. It has:

*   **Attributes (Data):** color, brand, model, year, current speed, fuel level.
*   **Behaviors (Methods):** `start_engine()`, `accelerate()`, `brake()`, `turn_left()`.

OOP allows us to create a "Car" blueprint (called a **Class**) in our code, and from that blueprint, we can create individual car objects, each with its own unique attributes (e.g., a red Ferrari, a blue Toyota).

### Why is OOP so popular and powerful?

1.  **Modularity:** It breaks down complex problems into smaller, self-contained, and manageable objects. This makes your code easier to read, understand, and maintain.
2.  **Reusability:** Once you create a class, you can reuse it to create multiple objects without rewriting the code. You can also reuse classes in other programs (e.g., a `Button` class can be used in many different applications).
3.  **Scalability:** OOP makes it easier to build large and complex applications. As your project grows, you can add new objects and features without breaking existing code.
4.  **Easier Debugging:** When an error occurs, it's often easier to pinpoint which object is causing the problem, rather than searching through a long script.
5.  **Data Protection (Encapsulation):** OOP allows you to hide the internal complexity of an object and only expose the necessary functionalities. This prevents accidental modification of data and makes your code more secure.

In procedural programming, you might have separate variables for `car_color`, `car_brand`, and functions like `start_car(car_details)`. In OOP, all of this is neatly bundled together into a single `Car` object.

## 3. The `__init__` Method (Constructor)

When we create an object, we usually want to initialize it with some starting data (attributes). The `__init__` method is a special method in Python classes that is automatically called whenever a new object of that class is created.

It's often called the **constructor** because it constructs the object by setting its initial state.

### Syntax and the `self` Parameter

The `__init__` method is defined like any other method, but with a special name. The first parameter of `__init__` (and any instance method) is always `self`.

*   `self`: Represents the instance of the object itself. It's how the object refers to its own attributes and methods. When you call a method on an object like `my_object.my_method()`, Python automatically passes `my_object` as the `self` argument.

```python
class Dog:
    def __init__(self, name, age):

        self.name = name
        self.age = age
        print(f"A new dog named {self.name} has been created!")

# Now, when we create a Dog object, we must provide the arguments for __init__
Dog("Buddy", 3)
Dog("Lucy", 5)
dog1=Dog("Buddy", 3)
dog2=Dog("Lucy", 5)
# We can access the attributes of each instance using dot notation
print(f"{dog1.name} is {dog1.age} years old.")
print(f"{dog2.name} is {dog2.age} years old.")
```

**Explanation:**
1.  When we write `dog1 = Dog("Buddy", 3)`, Python creates a new `Dog` object.
2.  It then calls the `__init__` method, passing the new object as `self`, `"Buddy"` as `name`, and `3` as `age`.
3.  Inside `__init__`, `self.name = name` creates an attribute `name` on the `dog1` object and assigns it the value `"Buddy"`.
4.  Similarly, `self.age = age` creates an attribute `age` and assigns it the value `3`.

## 4. Instance Attributes and Methods

### Instance Attributes

**Instance attributes** are variables that belong to a specific instance of a class. Each object has its own copy of these attributes. We created instance attributes in the previous section using `self.attribute_name = value` inside the `__init__` method.

### Instance Methods

**Instance methods** are functions defined inside a class that can access and modify the instance's attributes. The first parameter of an instance method is always `self`, which allows the method to work with the specific object's data.

### Analogy: A Smartphone

*   **Instance Attributes:** Your specific phone's serial number, color, storage capacity, and the apps you have installed.
*   **Instance Methods:** Actions you can perform on your phone, like `make_call(phone_number)`, `send_text(message)`, `install_app(app_name)`. These actions use or change your phone's specific attributes (e.g., `make_call` uses your phone's SIM card).

### Example: Adding Methods to our `Dog` Class

Let's add some behaviors (methods) to our `Dog` class.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"\n name = {self.name}")

    # This is an instance method
    def bark(self):
        print(f"{self.name} says: Woof! Woof!")

    # Another instance method that uses an attribute
    def get_human_age(self):
        # human_age = self.age * 7
        print( f"{self.name} is {self.age * 7} in human years.")

# Create an instance
dog1 = Dog("Buddy", 3)

# Call the instance methods on the object
dog1.bark()
dog1.get_human_age()

# Create another instance
# dog2 = Dog("Lucy", 5)
# dog2.bark()
# print(dog2.get_human_age())
```

**Explanation:**
*   `bark()` and `get_human_age()` are instance methods. They both take `self` as their first parameter, allowing them to access the instance's data (`self.name`, `self.age`).
*   When we call `dog1.bark()`, Python passes `dog1` as the `self` argument to the `bark` method.

## 5. Class Attributes and Methods (`@classmethod`)

So far, we've dealt with **instance attributes** (data unique to each object) and **instance methods** (behaviors unique to each object). But what if we want data or behavior that belongs to the *class itself*, rather than to any specific instance?

This is where **class attributes** and **class methods** come in.

### Class Attributes

**Class attributes** are variables that are shared by all instances of a class. They are defined directly inside the class, but outside of any method.

### Analogy: Car Manufacturer

Imagine a car manufacturer. Each car (instance) has its own color, model, and VIN (instance attributes). But the *manufacturer* has a common `company_name` or a `number_of_wheels` that applies to all cars they produce. These would be class attributes.

```python
class Car:
    # Class attribute: shared by all instances of Car
    number_of_wheels = 4
    manufacturer = "Generic Motors"

    def __init__(self, model, color):
        self.model = model  # Instance attribute
        self.color = color  # Instance attribute

# Accessing class attributes using the class name
print(f"All cars have {Car.number_of_wheels} wheels.")
print(f"Manufacturer: {Car.manufacturer}")

# Accessing class attributes using an instance (though less common for modification)
my_car = Car("Sedan", "Blue")
another_car = Car("SUV", "Red")

print(f"My car has {my_car.number_of_wheels} wheels.")
print(f"Another car is from {another_car.manufacturer}.")

# If you change a class attribute via the class, it affects all instances
Car.number_of_wheels = 6 # Imagine a new type of vehicle
print(f"Now all cars have {my_car.number_of_wheels} wheels (after change).")
```

**Explanation:**
*   `number_of_wheels` and `manufacturer` are defined directly under `class Car:`. They are class attributes.
*   Changes to `Car.number_of_wheels` affect all instances because they all refer to the same shared attribute.

### Class Methods (`@classmethod`)

**Class methods** are methods that operate on the class itself, rather than on an instance of the class. They receive the class as their first argument (conventionally named `cls`), instead of the instance (`self`).

They are defined using the `@classmethod` decorator.

### Analogy: Factory Settings

Think of a factory that produces cars. A class method might be `Car.set_default_color("Green")`. This method doesn't operate on a specific car, but rather changes a setting for *all* cars produced by that factory (the class).

## 6. Static Methods (`@staticmethod`)

**Static methods** are methods that belong to the class but do not operate on the class or its instances. They don't receive `self` or `cls` as their first argument. They are essentially regular functions that are logically grouped within a class because they have some conceptual relationship with the class.

They are defined using the `@staticmethod` decorator.

### Analogy: Utility Functions in a Toolbox

Imagine a toolbox (your class) that contains various tools. Some tools (instance methods) need a specific object to work on (e.g., a wrench needs a nut). Some tools (class methods) affect the entire toolbox (e.g., a tool to organize all tools). A static method is like a general-purpose utility in the toolbox, like a measuring tape. It's useful, it's in the toolbox, but it doesn't need to know anything about the specific toolbox or any specific item in it to function.

```python
class MathOperations:
    def __init__(self, value):
        self.value = value

    def instance_method(self):
        print(f"Instance method called on value: {self.value}")

    @classmethod
    def class_method(cls):
        print(f"Class method called on class: {cls.__name__}")

    @staticmethod
    def add(x, y):
        """Adds two numbers. Does not need instance or class data."""
        return x + y

    @staticmethod
    def is_even(number):
        """Checks if a number is even."""
        return number % 2 == 0

# Calling static methods using the class name (recommended)
print(f"5 + 3 = {MathOperations.add(5, 3)}")
print(f"Is 4 even? {MathOperations.is_even(4)}")
print(f"Is 7 even? {MathOperations.is_even(7)}")

# You can also call them via an instance, but it's less intuitive
calc = MathOperations(10)
print(f"10 + 20 = {calc.add(10, 20)}")

# Demonstrate other methods for comparison
calc.instance_method()
MathOperations.class_method()
```

**Explanation:**
*   `add` and `is_even` are static methods. They don't take `self` or `cls`.
*   They perform operations that are related to the class's purpose (mathematics) but don't require any specific instance data or class-level state.

### When to use Static Methods?

*   **Utility functions:** When a function logically belongs to a class but doesn't need to access any instance or class-specific data.
*   **Helper functions:** Functions that assist other methods in the class but don't depend on the object's state.
*   **No `self` or `cls` needed:** If your method doesn't use `self` or `cls` at all, it's a good candidate for a static method.

## 7. Encapsulation (Public, Protected, Private)

**Encapsulation** is one of the fundamental principles of OOP. It refers to the bundling of data (attributes) and methods (functions) that operate on the data into a single unit (a class). It also involves restricting direct access to some of an object's components, meaning that the internal representation of an object is hidden from the outside.

### Analogy: A TV Remote Control

Think of a TV remote control. You use buttons like "Volume Up," "Channel Down," or "Power." You don't need to know *how* the remote changes the volume (the internal circuitry and signals). You just interact with its public interface (the buttons). The internal workings are encapsulated.

In Python, encapsulation is achieved through conventions, as Python doesn't have strict access modifiers like `public`, `private`, or `protected` as found in some other languages (e.g., Java, C++).

### Access Modifiers (Conventions in Python):

1.  **Public Members (No underscore):**
    *   Attributes and methods without any leading underscores are considered public.
    *   They can be accessed and modified from anywhere, both inside and outside the class.
    *   **Convention:** You are free to use them directly.

```python
class MyClass:
        def __init__(self):
            self.public_attribute = "I am public"

        def public_method(self):
            print("This is a public method.")

obj = MyClass()
print(obj.public_attribute)
obj.public_method()

obj.public_attribute = "I can be changed"
print(obj.public_attribute)
```

2.  **Protected Members (Single leading underscore `_`):**
    *   Attributes and methods with a single leading underscore (e.g., `_protected_attribute`) are considered protected.
    *   **Convention:** They should not be accessed directly from outside the class, but they can be accessed by subclasses (which we'll cover in Inheritance).
    *   Python does *not* strictly enforce this; it's a hint to other developers that this is an internal part of the class and should be treated with care.

```python
class MyClass:
        def __init__(self):
            self._protected_attribute = "I am protected"

        def _protected_method(self):
            print("This is a protected method.")

obj = MyClass()
# You can still access them, but it's discouraged by convention
print(obj._protected_attribute)
obj._protected_method()
```

3.  **Private Members (Double leading underscore `__`):**
    *   Attributes and methods with a double leading underscore (e.g., `__private_attribute`) are considered private.
    *   Python *does* perform a name mangling process for these. This means the name of the attribute/method is changed to `_ClassName__private_attribute`.
    *   This makes it harder (but not impossible) to access them directly from outside the class, providing a stronger form of encapsulation.
    *   **Convention:** They are intended for internal use within the class only.

```python
class MyClass:
        def __init__(self):
            self.__private_attribute = "I am private"

        def __private_method(self):
            print("This is a private method.")

        def get_private_attribute(self):
            return self.__private_attribute

obj = MyClass()
    # print(obj.__private_attribute) # This will raise an AttributeError

    # Accessing private members via a public method (the intended way)
print(obj.get_private_attribute())

    # You can still access it if you know the mangled name (discouraged!)
print(obj._MyClass__private_attribute)
```

**Summary of Encapsulation in Python:**

| Convention          | Prefix      | Enforcement | Purpose                                    |
| :------------------ | :---------- | :---------- | :----------------------------------------- |
| **Public**          | None        | None        | Accessible from anywhere                   |
| **Protected**       | `_` (single)| By convention | Internal use, accessible by subclasses     |
| **Private**         | `__` (double)| Name mangling | Strictly internal to the class             |

## 8. Getters and Setters (`@property`)

While direct access to public attributes is common in Python, sometimes you want more control over how attributes are accessed or modified. This is where **getters** and **setters** come in. They are methods that provide controlled access to an object's attributes.

Python provides a convenient way to create getters and setters using the `@property` decorator.

### Why use Getters and Setters?

*   **Validation:** You can add logic to a setter to validate the data before it's assigned to an attribute (e.g., ensuring an age is positive).
*   **Computed Attributes:** A getter can compute a value on the fly rather than storing it directly (e.g., `full_name` from `first_name` and `last_name`).
*   **Encapsulation:** They allow you to change the internal representation of an attribute without affecting the external interface that uses it.
*   **Logging/Side Effects:** You can log when an attribute is accessed or modified, or trigger other actions.

### The `@property` Decorator

Python's `@property` decorator allows you to define methods that can be accessed like attributes. This is often called a "Pythonic" way to implement getters and setters.

```python
class Person:
    def __init__(self, name, age):
        self._name = name # Use a protected attribute for internal storage
        self._age = age

    # Getter for name
    @property
    def name(self):
        print("Getting name...")
        return self._name

    # Setter for name
    @name.setter
    def name(self, new_name):
        print("Setting name...")
        if not isinstance(new_name, str) or len(new_name) == 0:
            raise ValueError("Name must be a non-empty string.")
        self._name = new_name

    # Getter for age
    @property
    def age(self):
        print("Getting age...")
        return self._age

    # Setter for age with validation
    @age.setter
    def age(self, new_age):
        print("Setting age...")
        if not isinstance(new_age, (int, float)) or new_age < 0:
            raise ValueError("Age must be a non-negative number.")
        self._age = new_age

# Create an instance
p = Person("Alice", 30)

# Accessing attributes using the getter (looks like direct attribute access)
print(f"Person's name: {p.name}")
print(f"Person's age: {p.age}")

# Modifying attributes using the setter (looks like direct attribute assignment)
p.name = "Bob"
p.age = 25
print(f"Updated name: {p.name}, Updated age: {p.age}")

# Try invalid assignments
try:
    p.age = -5
except ValueError as e:
    print(f"Error: {e}")

try:
    p.name = ""
except ValueError as e:
    print(f"Error: {e}")
```

**Explanation:**
*   We define a method `name` and decorate it with `@property`. This makes `name` an attribute that can be read.
*   To make it writable, we define another method with the same name (`name`) and decorate it with `@name.setter`. This method will be called when you try to assign a value to `p.name`.
*   Inside the setter, we can add validation logic.

### Read-Only Properties

If you only define the `@property` getter and do not define a corresponding `@attribute_name.setter`, the property becomes read-only.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @property
    def area(self):
        return 3.14159 * self._radius ** 2

c = Circle(5)
print(f"Circle radius: {c.radius}")
print(f"Circle area: {c.area:.2f}")
try:

 c.radius = 6 # This would raise an AttributeError because no setter is defined
 c.area = 100 # This would also raise an AttributeError
except AttributeError:
  print ("error")
```

*   **Extensibility:** You can extend the functionality of an existing class without modifying it. Child classes can add new attributes and methods or override existing ones.
*   **Maintainability:** Changes in the parent class automatically propagate to child classes, simplifying maintenance.
*   **Logical Grouping:** It helps in organizing classes into a logical hierarchy, making the code more structured and understandable.

## 10. Multiple Inheritance

**Multiple inheritance** is a feature in Python (and some other OOP languages) where a class can inherit from more than one parent class. This allows a child class to combine the functionalities of multiple base classes.

### Analogy: Hybrid Creatures

Imagine a mythical creature that is part bird and part fish. It inherits the ability to fly from the bird and the ability to swim from the fish. This creature is a good analogy for a class that inherits from multiple parent classes.

### Syntax for Multiple Inheritance

To inherit from multiple classes, you list them in parentheses, separated by commas, after the child class name.

```python
class Parent1:
    def method1(self):
        print("Method from Parent1")

class Parent2:
    def method2(self):
        print("Method from Parent2")

class Child(Parent1, Parent2):
    def method_child(self):
        print("Method from Child")

# Create an instance of the child class
obj = Child()
obj.method1()
obj.method2()
obj.method_child()
```

### Method Resolution Order (MRO)

When a method is called on an object that inherits from multiple classes, Python needs a way to determine which parent class method to execute if there are methods with the same name in different parent classes. This is handled by the **Method Resolution Order (MRO)**.

Python uses a C3 linearization algorithm to determine the MRO. You can inspect the MRO of a class using `ClassName.__mro__` or `help(ClassName)`.

```python
class A:
    def hi(self):
        print("A")

class B(A):
    def hi(self):
        print("B")

class C(A):
    def hi(self):
        print("C")

class D(B, C):
    def hi(self):
        print("D")

d = D()
d.hi() # Which hi() will be called?

print(D.__mro__)
```

**Explanation:**
*   In the example above, `D` inherits from `B` and `C`. Both `B` and `C` inherit from `A`. All classes have a `hi()` method.
*   Python follows the MRO to find the method. The MRO for `D` is `(D, B, C, A, object)`. So, `D.hi()` is called first, then `B.hi()`, then `C.hi()`, then `A.hi()`.

### Example: `Flyable` and `Swimmable` Mixins

Mixins are classes that provide a specific functionality to be inherited by other classes. They are not meant to be instantiated on their own.

```python
class Flyable:
    def fly(self):
        print("I can fly!")

class Swimmable:
    def swim(self):
        print("I can swim!")

class Duck(Flyable, Swimmable):
    def __init__(self, name):
        self.name = name

    def quack(self):
        print(f"{self.name} says Quack!")

duck = Duck("Donald")
duck.quack()
duck.fly()
duck.swim()

print(Duck.__mro__)
```

## 11. Method Overriding and `super()`

**Method Overriding** is a feature that allows a child class to provide a specific implementation for a method that is already defined in its parent class. When a method is overridden, the version in the child class is executed instead of the parent class version when called on an object of the child class.

### Analogy: Customizing a Recipe

Imagine you have a basic cookie recipe (parent class method). You can use it as is, or you can *override* it to create a chocolate chip cookie recipe (child class method) by adding chocolate chips to the basic steps. The core idea of making a cookie is still there, but you've customized it.

### Syntax for Method Overriding

Simply define a method in the child class with the exact same name as a method in the parent class.

```python
class Animal:
    def speak(self):
        print("Animal makes a sound.")

class Dog(Animal):
    def speak(self):
        print("Dog barks: Woof!")

class Cat(Animal):
    def speak(self):
        print("Cat meows: Meow!")

# Create instances
animal = Animal()
dog = Dog()
cat = Cat()

animal.speak()
dog.speak() # Dog's speak() is called, overriding Animal's
cat.speak() # Cat's speak() is called, overriding Animal's
```

### The `super()` Function

The `super()` function provides a way to call a method from the parent class (superclass) within a child class. This is particularly useful when you want to extend the functionality of a parent method rather than completely replacing it.

### Analogy: Building on Foundations

If you're building a house, you first lay the foundation (parent class). Then, you build the walls (child class). When building the walls, you still need to refer to the foundation. `super()` is like saying, "First, do what the parent class does, then add my own specific stuff."

### Using `super()` in `__init__`

The most common use of `super()` is in the `__init__` method of a child class to call the `__init__` method of its parent class. This ensures that the parent class's attributes are properly initialized.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"Hi, my name is {self.name} and I am {self.age} years old.")

class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age) # Call parent's __init__
        self.student_id = student_id

    def introduce(self):
        super().introduce() # Call parent's introduce() method
        print(f"I am also a student with ID: {self.student_id}.")

# Create instances
p = Person("Alice", 30)
p.introduce()

s = Student("Bob", 20, "S12345")
s.introduce()
```

**Explanation:**
*   In `Student.__init__`, `super().__init__(name, age)` calls the `Person` class's `__init__` to handle the initialization of `name` and `age`.
*   In `Student.introduce`, `super().introduce()` calls the `Person` class's `introduce` method, and then the `Student` class adds its own specific message.

## 12. Polymorphism in Python

**Polymorphism** (from Greek, meaning "many forms") is the ability of an object to take on many forms. In OOP, it refers to the ability of different classes to be treated as instances of a common interface or superclass, even if they implement the common methods in different ways.

### Analogy: A Remote Control for Different Devices

Imagine a universal remote control. You can use the "Play" button on the remote, and it will work differently depending on whether you are controlling a DVD player, a streaming device, or a gaming console. The "Play" action (method) takes many forms depending on the object it's applied to.

### How Polymorphism Works in Python

Python is inherently polymorphic because it is a dynamically typed language. This means that Python doesn't care about the explicit type of an object as long as it supports the operation being performed. This concept is often referred to as **"Duck Typing"**:

> "If it walks like a duck and quacks like a duck, then it must be a duck."

This means if an object has the methods or attributes that a particular operation expects, then it can be used in that operation, regardless of its actual class.

### Example: Polymorphism with Functions

```python
class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

class Duck:
    def speak(self):
        return "Quack!"

def make_animal_speak(animal):
    print(animal.speak())

# Create instances of different classes
dog = Dog()
cat = Cat()
duck = Duck()

# Call the same function with different objects
make_animal_speak(dog)
make_animal_speak(cat)
make_animal_speak(duck)
```

**Explanation:**
*   The `make_animal_speak` function doesn't care if the `animal` object is a `Dog`, `Cat`, or `Duck`. It only cares that the `animal` object has a `speak()` method.
*   This is polymorphism in action: the same method call (`animal.speak()`) behaves differently depending on the object it's called on.

### Example: Polymorphism with Inheritance

Polymorphism is often used in conjunction with inheritance. A parent class defines a common interface (method), and child classes provide their own specific implementations.

```python
class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement this method")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius ** 2

def print_area(shape):
    print(f"The area is: {shape.area():.2f}")

# Create instances of different shapes
rect = Rectangle(10, 5)
circ = Circle(7)

# Call the same function with different shape objects
print_area(rect)
print_area(circ)
```

**Explanation:**
*   `Shape` defines an `area()` method, acting as an interface.
*   `Rectangle` and `Circle` both implement their own versions of `area()`.
*   The `print_area` function can accept any object that has an `area()` method, demonstrating polymorphism.

## 14. Composition vs. Inheritance

When designing your classes, you often face a choice: should one class **inherit** from another, or should it **contain** an instance of another class? This is the core of the "Composition over Inheritance" principle.

### Inheritance (Is-A Relationship)

*   **Concept:** Inheritance models an "is-a" relationship. A `Dog` *is an* `Animal`. A `Car` *is a* `Vehicle`.
*   **Mechanism:** The child class gets all the attributes and methods of the parent class.
*   **Pros:** Code reuse, clear hierarchy, polymorphism.
*   **Cons:** Tight coupling (child is highly dependent on parent), can lead to complex hierarchies (diamond problem in multiple inheritance), difficult to change behavior at runtime.

### Composition (Has-A Relationship)

*   **Concept:** Composition models a "has-a" relationship. A `Car` *has an* `Engine`. A `Computer` *has a* `CPU` and `RAM`.
*   **Mechanism:** One class contains an instance of another class as an attribute.
*   **Pros:** Loose coupling (classes are independent), flexible (behavior can be changed at runtime by swapping components), easier to test.
*   **Cons:** Can require more code to delegate methods.

### Analogy: Building a Computer

*   **Inheritance:** If you were to use inheritance, you might say a `GamingComputer` *is a* `Computer`. This works for some aspects, but what if you want to swap out the `CPU`? You can't easily change the fundamental type of a `GamingComputer` at runtime.
*   **Composition:** A `Computer` *has a* `CPU`, *has* `RAM`, *has a* `StorageDevice`. You can build a `Computer` by assembling these separate components. If you want to upgrade the `CPU`, you just replace the `CPU` object within the `Computer` object.

### Example: Inheritance (`Vehicle` and `Car`)

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"{self.brand} vehicle started.")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def drive(self):
        print(f"{self.brand} {self.model} is driving.")

my_car = Car("Toyota", "Camry")
my_car.start()
my_car.drive()
```

### Example: Composition (`Engine` and `Car`)

```python
class Engine:
    def __init__(self, fuel_type):
        self.fuel_type = fuel_type

    def start(self):
        print(f"{self.fuel_type} engine started.")

class Car:
    def __init__(self, brand, model, engine):
        self.brand = brand
        self.model = model
        self.engine = engine # Car has an Engine object

    def drive(self):
        print(f"{self.brand} {self.model} is driving.")

    def start_car(self):
        self.engine.start() # Delegate the start behavior to the Engine object

# Create an Engine object
gas_engine = Engine("Gasoline")
electric_engine = Engine("Electric")

# Create Car objects with different engines
car1 = Car("Honda", "Civic", gas_engine)
car2 = Car("Tesla", "Model 3", electric_engine)

car1.start_car()
car1.drive()

car2.start_car()
car2.drive()
```

**When to choose which?**

*   **Inheritance:** Use when there is a clear "is-a" relationship and you want to reuse implementation and establish a strong type hierarchy.
*   **Composition:** Use when there is a "has-a" relationship, when you want flexibility, loose coupling, and the ability to change behavior at runtime.

**General Rule of Thumb:** Favor composition over inheritance when possible. It often leads to more flexible and maintainable designs.
