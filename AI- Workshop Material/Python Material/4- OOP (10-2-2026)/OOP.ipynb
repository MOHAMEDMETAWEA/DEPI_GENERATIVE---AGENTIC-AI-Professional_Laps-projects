## 1. Encapsulation

**Definition:** Encapsulation is the principle of bundling data (attributes) and methods (functions) within a class, while restricting access to the internal details. It protects data from external interference and misuse by using private attributes (prefixed with `_` or `__`) and providing public methods (getters/setters) for access.

**Benefits:** Improves security, maintainability, and modularity.

**Examples:** We'll show 5 examples demonstrating encapsulation with private attributes and controlled access.

### Example 1: Basic Encapsulation with Private Attribute

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age  # Private attribute (by convention)
    
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age must be positive.")

p = Person("Alice", 30)
print(p.get_age())  # Output: 30
p.set_age(35)
print(p.get_age())  # Output: 35
p.set_age(-5)  # Output: Age must be positive.
```

### Example 2: Encapsulation in Bank Account (Balance Protection)

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Strongly private
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    
    def get_balance(self):
        return self.__balance

acc = BankAccount(100)
acc.deposit(50)
print(acc.get_balance())  # Output: 150
# print(acc.__balance)  # Error: attribute not accessible
```

### Example 3: Encapsulation in Car Class (Speed Limit)

```python
class Car:
    def __init__(self, speed):
        self._speed = speed
    
    def accelerate(self, increase):
        self._speed += increase
        if self._speed > 200:
            self._speed = 200  # Enforce limit
     
    def get_speed(self):
        return self._speed

c = Car(100)
c.accelerate(150)
print(c.get_speed())  # Output: 200 (limited)
```

### Example 4: Encapsulation with Password Protection

```python
class SecureData:
    def __init__(self, data, password):
        self.__data = data
        self.__password = password
    
    def access_data(self, pwd):
        if pwd == self.__password:
            return self.__data
        else:
            return "Access denied"

sd = SecureData("Secret info", "pass123")
print(sd.access_data("passssss"))  # Output: Access denied
print(sd.access_data("pass123"))  # Output: Secret info
```

## 2. Inheritance

**Definition:** Inheritance allows a class (child/subclass) to inherit attributes and methods from another class (parent/superclass), promoting code reuse. The child can extend or override parent behavior.

**Benefits:** Reduces duplication, enables hierarchy (e.g., Animal → Dog).

**Examples:** 5 examples showing inheritance, super(), overriding.

### Example 1: Basic Inheritance (Animal → Dog)

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"

class Dog(Animal):
    def speak(self):
        return "Woof!"

d = Dog("Buddy")
print(d.name)  # Output: Buddy
print(d.speak())  # Output: Woof!
```

### Example 2: Inheritance with super() (Vehicle → Car)

```python
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

class Car(Vehicle):
    def __init__(self, make, model, doors):
        super().__init__(make, model)
        self.doors = doors

c = Car("Toyota", "Corolla", 4)
print(c.make, c.model, c.doors)  # Output: Toyota Corolla 4
```

### Example 3: Method Overriding in Inheritance (Shape → Rectangle)

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

r = Rectangle(5, 3)
print(r.area())  # Output: 15
```

### Example 4: Multiple Inheritance (Employee → Manager)

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Person:
    def __init__(self, age):
        self.age = age

class Manager(Employee, Person):
    def __init__(self, name, salary, age, department):
        Employee.__init__(self, name, salary)
        Person.__init__(self, age)
        self.department = department

m = Manager("Alice", 80000, 40, "HR")
print(m.name, m.salary, m.age, m.department)  # Output: Alice 80000 40 HR
```

### Example 5: Inheritance Chain (Animal → Mammal → Human)

```python
class Animal:
    def move(self):
        return "Moves"

class Mammal(Animal):
    def feed(self):
        return "Feeds milk"

class Human(Mammal):
    def speak(self):
        return "Talks"

h = Human()
print(h.move())  # Output: Moves
print(h.feed())  # Output: Feeds milk
print(h.speak())  # Output: Talks
```

## 3. Polymorphism

**Definition:** Polymorphism allows different classes to be treated as instances of the same class through a common interface. It enables method overriding and operator overloading, so the same method name can behave differently based on the object.

**Benefits:** Flexible code, easy to extend with new classes.

**Examples:** 5 examples showing polymorphism via overriding and common interfaces.

### Example 1: Polymorphism with Method Overriding (Shapes)

```python
class Shape:
    def area(self):
        return 0

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius ** 2

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side ** 2

shapes = [Circle(5), Square(4)]
for s in shapes:
    print(s.area())  # Output: 78.5 and 16 (different behaviors)
```

### Example 2: Polymorphism in Animals Speak

```python
class Animal:
    def speak(self):
        return "Sound"

class Dog(Animal):
    def speak(self):
        return "Woof"

class Cat(Animal):
    def speak(self):
        return "Meow"

animals = [Dog(), Cat()]
for a in animals:
    print(a.speak())  # Output: Woof and Meow
```

### Example 3: Operator Overloading (Polymorphism with +)

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1 + p2)  # Output: (4, 6)
```

### Example 4: Polymorphism with Functions (len())

```python
class CustomList:
    def __init__(self, items):
        self.items = items
    
    def __len__(self):
        return len(self.items) * 2  # Custom behavior

cl = CustomList([1, 2, 3])
print(len(cl))  # Output: 6 (polymorphic len())
```

### Example 5: Polymorphism in Printing (str overriding)

```python
class Book:
    def __init__(self, title):
        self.title = title
    def __str__(self):
        return f"Book: {self.title}"

class Movie:
    def __init__(self, title):
        self.title = title
    def __str__(self):
        return f"Movie: {self.title}"

media = [Book("Python 101"), Movie("Inception")]
for m in media:
    print(m)  # Output: Book: Python 101 and Movie: Inception
```

## 4. Abstraction

**Definition:** Abstraction hides the complex implementation details and shows only the essential features. In Python, it's achieved using abstract classes and methods (via ABC module), forcing subclasses to implement specific methods.

**Benefits:** Ensures consistency and simplifies user interaction.

**Examples:** 4 examples using ABC for abstraction.

### Example 1: Abstract Shape Class

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height

t = Triangle(10, 5)
print(t.area())  # Output: 25.0
```

### Example 2: Abstract Payment Method

```python
class Payment(ABC):
    @abstractmethod
    def process(self, amount):
        pass

class CreditCard(Payment):
    def process(self, amount):
        return f"Paid {amount} via Credit Card"

class PayPal(Payment):
    def process(self, amount):
        return f"Paid {amount} via PayPal"

payments = [CreditCard(), PayPal()]
for p in payments:
    print(p.process(100))
```

### Example 3: Abstract Employee Role

```python
class Employee(ABC):
    @abstractmethod
    def work(self):
        pass

class Developer(Employee):
    def work(self):
        return "Writing code"

class Designer(Employee):
    def work(self):
        return "Creating designs"

team = [Developer(), Designer()]
for member in team:
    print(member.work())
```

### Example 4: Abstract Data Processor

```python
class DataProcessor(ABC):
    @abstractmethod
    def process_data(self, data):
        pass

class CSVProcessor(DataProcessor):
    def process_data(self, data):
        return f"Processed CSV: {len(data)} rows"

csv_p = CSVProcessor()
print(csv_p.process_data([[1,2],[3,4]]))  # Output: Processed CSV: 2 rows
```

## Static Methods and Class Methods

**Static Methods:**
- Defined with `@staticmethod`
- Don't access class or instance data (no self or cls)
- Like regular functions but inside class for organization
- Call via class or instance

**Class Methods:**
- Defined with `@classmethod`
- Receive cls as first argument
- Access/modify class state
- Useful for alternative constructors
- Call via class or instance

**Examples:** We'll show examples for both.

### Static Method Example: Utility Function

```python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

print(MathUtils.add(5, 3))  # Output: 8 (no instance needed)
```

### Class Method Example: Alternative Constructor

```python
class Date:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day
    
    @classmethod
    def from_string(cls, date_str):
        year, month, day = map(int, date_str.split('-'))
        return cls(year, month, day)

d = Date.from_string("2023-10-15")
print(d.year, d.month, d.day)  # Output: 2023 10 15
```

## Comprehensive Bank System Application

This application integrates all OOP principles:
- **Encapsulation**: Private balances, getters/setters.
- **Inheritance**: SavingsAccount and CheckingAccount inherit from BankAccount.
- **Polymorphism**: Different accounts override withdraw() behavior.
- **Abstraction**: Abstract BankAccount forces subclasses to implement methods.
- Static/Class Methods: Utility for account number generation, class method for stats.

System features:
- Create savings/checking accounts
- Deposit/withdraw with rules
- Interest for savings
- Overdraft for checking
- Track all accounts

```python
from abc import ABC, abstractmethod
import random

class BankAccount(ABC):
    accounts = []  # Class-level list to track all accounts

    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance  # Encapsulated
        self.account_number = self._generate_account_number()  # Encapsulated generation
        BankAccount.accounts.append(self)
    
    @staticmethod
    def _generate_account_number():
        return random.randint(100000, 999999)
    
    @classmethod
    def total_accounts(cls):
        return len(cls.accounts)
    
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited {amount}. New balance: {self._balance}")
        else:
            print("Invalid deposit amount.")
    
    @abstractmethod
    def withdraw(self, amount):
        pass  # Must be implemented by subclasses
    
    def get_balance(self):
        return self._balance

class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0, interest_rate=0.05):
        super().__init__(owner, balance)
        self.interest_rate = interest_rate
    
    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew {amount}. New balance: {self._balance}")
        else:
            print("Insufficient funds or invalid amount.")
    
    def add_interest(self):
        interest = self._balance * self.interest_rate
        self._balance += interest
        print(f"Added interest {interest}. New balance: {self._balance}")

class CheckingAccount(BankAccount):
    def __init__(self, owner, balance=0, overdraft_limit=100):
        super().__init__(owner, balance)
        self.overdraft_limit = overdraft_limit
    
    def withdraw(self, amount):
        if 0 < amount <= self._balance + self.overdraft_limit:
            self._balance -= amount
            print(f"Withdrew {amount}. New balance: {self._balance}")
        else:
            print("Overdraft limit exceeded or invalid amount.")

# Usage Example
savings = SavingsAccount("Alice", 1000)
savings.deposit(500)
savings.withdraw(200)
savings.add_interest()

checking = CheckingAccount("Bob", 500)
checking.deposit(300)
checking.withdraw(700)  # Uses overdraft

print(f"Total accounts: {BankAccount.total_accounts()}")
```
