```python
class Father:
    def __init__(self, name):
        self.name=name

    def sleep(self):
        return ('seelp on right')
    
class Mother:
    def __init__(self,name):
        self.name=name

    def sleep(self):
        return ('sleep on left')
    
class Boy(Father, Mother):
    def __init__(self, name):
        self.name=name

    def sleep(self, hours):
        return (f'sleep on middle and for {hours}')
    
    def sleep_(self):
        return ('sleep on left')
```

```python
boy1=Boy('Ahmed')
boy1.sleep(4)
```
