```python
# Practical guard + all()/any()
def validate_user(age, name, scores):
    if not name:
        return "Invalid name"
    if not all(s >= 0 for s in scores):
        return "Negative score"
    if any(s > 100 for s in scores):
        return "Score too high"
    return "Valid" if age >= 18 else "Underage"

print(validate_user(20, "Bob", [85, 95, 102]))
```

## 3. Loops

**Important patterns**
- `for ... in ...`
- `while`
- `break` / `continue` / `else` clause
- `enumerate()`, `zip()`
- **List / dict / set comprehensions** (very idiomatic in Python)

```python
# enumerate + zip
names = ["Alice", "Bob", "Charlie"]
ages  = [30, 25, 35]

for i, (name, age) in enumerate(zip(names, ages), 1):
    print(f"#{i}: {name} → {age} years")
```

```python
# Loop + else (runs only if NO break happened)
num = 17
for i in range(2, num):
    if num % i == 0:
        print("Not prime")
        break
else:
    print("Prime!")
```

### List Comprehensions – Advanced Examples

```python
# Flatten nested list
nested = [[1,2,3], [4,5], [6]]
flat = [x for sub in nested for x in sub]
print(flat)
```

```python
# Transpose matrix (classic interview question)
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(transpose)
```

```python
# Primes (one-liner style – readable version)
primes = [n for n in range(2, 50) if all(n % i != 0 for i in range(2, n))]
print(primes)
```

## 4. Functions

**Key concepts**
- `*args`, `**kwargs`
- Default arguments (careful with mutable defaults!)
- Lambda functions
- Higher-order functions (functions that take/return functions)

```python
def describe_person(name, age=30, *hobbies, **details):
    print(f"{name}, {age} years old")
    if hobbies:
        print("Hobbies:", ", ".join(hobbies))
    for k, v in details.items():
        print(f"  {k}: {v}")

describe_person("Sara", 28, "reading", "yoga", city="Cairo", job="Data Analyst")
```

```python
# Lambda + sorted + map + filter
data = [("apple", 3), ("kiwi", 1), ("banana", 5)]

sorted_by_count = sorted(data, key=lambda x: x[1], reverse=True)
print(sorted_by_count)

squared_evens = list(map(lambda x: x**2, filter(lambda x: x % 2 == 0, range(1,11))))
print(squared_evens)
```
