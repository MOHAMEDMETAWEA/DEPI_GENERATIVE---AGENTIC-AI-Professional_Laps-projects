```python
class Passengers:
    def __init__(self, passenger_id, name):
        self.id=passenger_id
        self.name=name


class Flight:
    def __init__(self, number, destination,Time, seats):
        self.number=number
        self.destination=destination
        self.Time=Time
        self.seats=seats
        self.passengers=[]
    
# flight1=Flight(1,'Cairo', 30)

    def book(self,passenger):
        if self.seats==0:
            print('NO seats')
            return
        
        for p in self.passengers:
            if p.id==passenger.id:
                print('Already Booked')
                return

        self.passengers.append(passenger)
        self.seats-=1
        print('Booking done')


    def cancel(self, passenger_id):
        for p in self.passengers:
            if p.id==passenger_id:
                self.passengers.remove(p)
                self.seats+=1
                print('Cancelling Done')
                return
        print('Not Founded')  


    def show(self):
        print(f'the flight number= {self.number}')
        print(f'the flight desdination= {self.destination}')
        print(f'the flight Time={self.Time}')
        print(f'the Avaliable seats= {self.seats}')
        
        if len(self.passengers) ==0:
            print('No passengers')
        else:
            print('All Passengers: ')
            for p in self.passengers:
                print(p.id,'--', p.name)

## Airplane System
flights=[]

def find_flight(number):
    for f in flights:
        if f.number==number:
            return f
    return None    

def add_flight():
    number=input('Flight Number=')
    destination=input('destination: ')
    seats=int(input('Seats'))
    Time=input('flight Time')
    new_flight=Flight(number,destination, Time, seats )
    flights.append(new_flight)
    print('Flight Added')


def book_ticket():
    number=input('flight number')
    ff=find_flight(number)
    if ff is None:
        print('flight Not founded')
        return 
    id=input('enter passerger id:')
    name=input('Enter Passenger Name:')
    passenger=Passengers(id, name)
    ff.book(passenger)



def Cancel_ticket():
    number=input('Flight number to cancel: ')
    ff=find_flight(number)
    if ff is None:
        print('flight Not founded')
        return 
    
    pid=input('Enter id')
    ff.cancel(pid)

def show_flights():
    for ff in flights:
        ff.show()

def cancel_flight():
    number=input('Flight number to cancel: ')
    ff=find_flight(number)
    if ff is None:
        print('flight Not founded')
        return
    flights.remove(ff)

while True:
    print('1. Add Flight')
    print('2. Book Ticket')
    print('3. cancel Ticket')
    print('4. show Flights')
    print('5. Cancel Flight')
    print('6. Exit')

    choice=input('enter your choice: ')
    if choice=='1':
        add_flight()

    elif choice=='2':
        book_ticket()
    elif choice=='3':
        Cancel_ticket()
    elif choice=='4':
        show_flights()
    elif choice=='5':
        cancel_flight()
    elif choice=='6':
        break

    else:
        print('invalid choice')
```
