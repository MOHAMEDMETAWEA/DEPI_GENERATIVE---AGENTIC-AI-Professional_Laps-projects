```python
Tasks=[]
while True:
    print(" 1. Add Task\n 2.Delete Task\n 3.Show Tasks\n 4.Exit\n")
    choice=input('select what you want')

    if choice== '1':
        Tasks.append(input('Task: '))
    
    elif choice=='2':
        task_del_num=int(input('enter the task number you want to delete:'))
        del Tasks[task_del_num -1]

    elif choice == '3':
        for i in Tasks:
            print(i)
    
    elif choice=='4':
        break
    else:
        print('invalid input')
```

```python
import random
def play_game_prs():
    chos=['paper', 'rock','scissor']
    score={'player': 0, 'computer':0}

    for i in range(3):
        player= input("paper/ rock/ scissor").strip().lower()
        computer=random.choice(chos)
        print(f' the computer select {computer}')

        if player == computer:
            print('Equel')

        elif (player == 'paper' and computer== 'rock' ) or (player == 'Scissor' and computer== 'paper') or(player == 'rock' and computer== 'Scissor'):
           score['player']+=1
           # score['player']=score['player'] +1
           print('player wins') 

        else:
            score['computer']+=1
            print('computer wins')

    print(f" the final score for player= {score['player']} and computer score= {score['computer']}")


play_game_prs()
```

```python
employees={}

while True:
    name=input('Enter the name of employee (or Exit)')
    if name.lower() == 'exit':
        break
    
    points=[]
    hours=[]
    for day in range(5):
      point=int(input(f'Enter the point in day {day}'))
      points.append(point)  
      hour=int(input(f'Enter the hours in day {day}'))  
      hours.append(hour)
      
      employees[name]={'hours':hours ,
          'points': points,
          'sum_points':sum(points),
          'sum_hours':sum(hours),
           'average_points': sum(hours)/5 }
      
for name, data in employees.items():
   print(f'{name}: {data}')
      
```
