```python
import re

def extract_numbers(text):
    # Extract all numbers, including those with commas, from the text
    numbers = re.findall(r'\b(?:\d{{1,3}}(?:,\d{{3}})*|\d+)(?:\.\d+)?\b', text.replace('$', ''))
    # Convert to float or int, handle potential commas
    return [float(n.replace(',', '')) if '.' in n else int(n.replace(',', '')) for n in numbers]

# The previous `amount_userQ` was specific to an 'amount'.
# We will now handle numerical extraction more generically within `fullsystem`.
```

```python
def fullsystem(userQ, policy_embeddings):
  user_numbers = extract_numbers(userQ)
  userQ_embedding = embedder.encode([userQ], normalize_embeddings=True)
  score = (userQ_embedding @ policy_embeddings.T).squeeze()

  # Get the top 5 candidates to find the best logical match
  top_idx = np.argsort(-score)[:5]

  best_match = None
  for i in top_idx:
    policy_full_text = df.iloc[i]['full_text']
    condition_str = str(df.iloc[i]['condition'])
    policy_name = df.iloc[i]['policy_name']
    rule_description = df.iloc[i]['rule_description']
    threshold_value = df.iloc[i]['threshold_value']

    eval_vars = {
        'amount': user_numbers[0] if user_numbers else None,
        'flight_hours': user_numbers[0] if user_numbers else None,
        'hotel_cost': user_numbers[0] if user_numbers else None,
        'termination_notice_days': user_numbers[0] if user_numbers else None,
        'payment_terms_days': user_numbers[0] if user_numbers else None,
        'years_since_last_replacement': user_numbers[0] if user_numbers else None,
        'true': True, 'false': False, 'executive': 'executive', 'international': 'international', 'regional': 'regional'
    }
    eval_vars['role'] = 'executive' if 'executive' in userQ.lower() else ''
    eval_vars['region'] = 'international' if 'international' in userQ.lower() else ('regional' if 'regional' in userQ.lower() else '')

    condition_to_eval = condition_str.replace(' AND ', ' and ').replace(' OR ', ' or ')

    try:
        # Check if condition is met
        if eval(condition_to_eval, {'__builtins__': {}}, eval_vars) == True:
            best_match = f"MATCHED POLICY: {policy_full_text}"
            break
    except:
        continue

  return best_match if best_match else "No exact policy match found for the criteria provided."
```

# ## Task 2: Handling `flight_hours` and `amount` with LLM vs. Coding

We have two primary ways to handle specific numerical and categorical parameters like `flight_hours` and `amount` within our policy matching system:

1.  **Coding Approach (using `fullsystem`):** This method involves explicitly extracting numerical values and keywords from the user's query and programmatically evaluating the `condition` strings of the policies. This is a precise, rule-based approach where the system determines if a condition is met based on a predefined logical expression and extracted data. The `fullsystem` function currently does this.

2.  **LLM Approach (using `get_llm_best_policy`):** This method leverages the language understanding capabilities of the LLM. It's less about programmatic evaluation of strict rules and more about the LLM's ability to semantically understand the user's intent and the policy text. The LLM is provided with the user query and a set of candidate policies (which contain `flight_hours` or `amount` details) and then selects the most relevant one based on its learned knowledge. The `get_llm_best_policy` function uses this method.

Let's demonstrate both with examples focusing on `flight_hours` and `amount`.

```python
def get_final_answer(userQ, policy_embeddings):
    # 1. Get programmatic evaluation from fullsystem
    top_results = fullsystem(userQ, policy_embeddings)

    # 2. Extract the primary matching result
    # We look for the first SUCCESS (True) or relevant DENIED (False)
    best_match = None
    for res in top_results:
        if res['condition_met'] == True:
            best_match = res
            break

    if best_match:
        return f"APPROVED: Based on the policy '{best_match['full_text'].split('|')[1].strip()}', your request is allowed. Requirement: {best_match['full_text'].split('|')[-2].strip()}"

    # If no direct 'True' match, look for a 'False' match in the same category
    for res in top_results:
        if res['condition_met'] == False:
             return f"DENIED: Based on the policy '{res['full_text'].split('|')[1].strip()}', the request does not meet the requirements. Rule: {res['full_text'].split('|')[2].strip()}"

    return "COULD NOT DETERMINE: Please provide more details (like an exact amount or your role) to check the policy requirements."
```
