# Optimized RAG System Implementation

```python
def retrieve_policies(user_query, embedder, policy_embeddings, df, top_k=5):
    """
    Retrieves the top K relevant policies using semantic search.
    """
    query_embedding = embedder.encode([user_query], normalize_embeddings=True)
    scores = (query_embedding @ policy_embeddings.T).squeeze()
    top_indices = np.argsort(-scores)[:top_k]
    
    results = []
    for idx in top_indices:
        results.append({
            "score": scores[idx],
            "text": df.iloc[idx]["full_text"],
            "details": df.iloc[idx].to_dict()
        })
    return results
```

```python
def answer_user_query(user_query, retrieved_policies, model, tokenizer):
    """
    Uses the LLM to answer the user query based on retrieved policies.
    """
    context = "\n".join([f"- {p['text']}" for p in retrieved_policies])
    
    prompt = f"""You are a policy assistant. Below are relevant company policies and a user query.
Use the provided policies to answer the user query accurately. If the query involves specific conditions (like amounts, dates, or roles), verify if they meet the policy requirements.

Policies:
{context}

User Query: {user_query}

Answer:"""
    
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    
    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        temperature=0.1,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
    
    response = tokenizer.decode(outputs[0][inputs['input_ids'].shape[-1]:], skip_special_tokens=True).strip()
    return response
```

```python
# Pipeline Integration and Testing
def rag_pipeline(query):
    retrieved = retrieve_policies(query, embedder, policy_embeddings, df)
    answer = answer_user_query(query, retrieved, model, tokenizer)
    return answer

# Test Cases
test_queries = [
    "I need to buy a high-end laptop for $4500. Who needs to approve this?",
    "I want to book an economy class flight for 5 hours.",
    "Can I share restricted data via email?",
    "What is the policy for a regional hotel if I spend $200?"
]

for q in test_queries:
    print(f"\nProcessing Query: {q}")
    result = rag_pipeline(q)
    print(f"Answer: {result}\n" + "="*50)
```

# Programmatic Evaluation System

```python
import re

def extract_numerical_vars(query):
    """
    Extracts potential numerical variables from the query.
    """
    # Extract numbers
    numbers = re.findall(r'\b(?:\d{1,3}(?:,\d{3})*|\d+)(?:\.\d+)?\b', query.replace('$', ''))
    extracted = [float(n.replace(',', '')) if '.' in n else int(n.replace(',', '')) for n in numbers]
    
    # Basic heuristic mapping to policy variables
    val = extracted[0] if extracted else 0
    
    return {
        'amount': val,
        'flight_hours': val,
        'hotel_cost': val,
        'termination_notice_days': val,
        'payment_terms_days': val,
        'years_since_last_replacement': val,
        'role': 'executive' if 'executive' in query.lower() else 'employee',
        'region': 'international' if any(x in query.lower() for x in ['international', 'abroad']) else 'regional',
        'vendor_approved': 'approved' in query.lower(),
        'software_approved': 'approved' in query.lower(),
        'software_legitimate': 'pirated' not in query.lower()
    }

def match_policy_programmatic(user_query, df):
    """
    Programmatically evaluates policies against a query using rule-based logic.
    """
    vars = extract_numerical_vars(user_query)
    matched = []
    
    for _, row in df.iterrows():
        cond = str(row['condition']).replace('AND', 'and').replace('OR', 'or').replace('executive', "'executive'")
        
        # Add annual_budget_limit from row if needed
        eval_vars = vars.copy()
        eval_vars['annual_budget_limit'] = row['threshold_value'] if not pd.isna(row['threshold_value']) else 2500
        
        try:
            if eval(cond, {'__builtins__': {}}, eval_vars):
                matched.append(row['full_text'])
        except:
            pass
            
    return matched

def evaluate_rag_response(query, llm_answer, df):
    """
    Evaluates the LLM response against programmatic ground truth.
    """
    programmatic_matches = match_policy_programmatic(query, df)
    
    # Simple check: Does the LLM answer mention the programmatically matched policy keywords?
    # In a real scenario, this could be more sophisticated (e.g., using another LLM as evaluator)
    
    status = "PASS" if any(m[:20].lower() in llm_answer.lower() for m in programmatic_matches) else "REVIEW NEEDED"
    
    if not programmatic_matches:
        status = "N/A (No programmatic match)"
        
    return {
        "query": query,
        "programmatic_result": programmatic_matches,
        "evaluation": status
    }
```

```python
# Integration Test with Evaluation
test_query = "I need to buy a high-end laptop for $4500. Who needs to approve this?"
llm_answer = rag_pipeline(test_query)
eval_result = evaluate_rag_response(test_query, llm_answer, df)

print(f"\nQuery: {test_query}")
print(f"LLM Answer: {llm_answer}")
print(f"Programmatic Match: {eval_result['programmatic_result']}")
print(f"Evaluation Status: {eval_result['evaluation']}")
```

```python
# Expanded Query Testing Suite
expanded_queries = [
    "I need to terminate a contract in 45 days.",
    "I want to book an economy class flight for 8 hours.",
    "Is it okay if I install pirated software for testing?",
    "I need to buy hardware that costs $1200.",
    "What is the policy for a hotel stay in an international city that costs $500?",
    "I am an executive and want to book a business class flight for 8 hours."
]

print(f"{'Query':<70} | {'LLM Match?':<12} | {'Prog Match?':<12} | {'Eval Status'}")
print("-"*120)

for q in expanded_queries:
    llm_ans = rag_pipeline(q)
    eval_res = evaluate_rag_response(q, llm_ans, df)
    
    # Simplified display logic
    llm_found = "Yes" if len(llm_ans) > 20 else "No"
    prog_found = "Yes" if eval_res['programmatic_result'] else "No"
    
    print(f"{q[:68]:<70} | {llm_found:<12} | {prog_found:<12} | {eval_res['evaluation']}")
```
