```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
       ("system", "انت مساعد عربي ذكي يساعد المستخدم في الإجابة على الأسئلة. وحافظ على سياق المحادثة"),
        ("human", "{user_question}"),
        ("placeholder", "{History}")
    ]
    
)
```

```python
prompt
```

```python
chat_chain= prompt | llm 
```

```python
# RunnableWithMessageHistory 


# InMemoryChatMessageHistory
# PostgresChatMessageHistory

from langchain_core.chat_history import InMemoryChatMessageHistory

store = {}
def _ensure_user(user_id):
    if user_id not in store:
        store[user_id] = {
            "LangChainHistory": InMemoryChatMessageHistory(),
            "prog_history": []
                
        }

def get_history(user_id):
    _ensure_user(user_id)
    return store[user_id]["LangChainHistory"]
```

```python
get_history("user1")
```

```python
from langchain_core.runnables.history import RunnableWithMessageHistory

chat_with_history = RunnableWithMessageHistory(
    chat_chain,
    get_history, 
    input_message_key="user_question",
    history_message_key="History"
)
```

```python
question = "what is the capital of France?"
user_id = "user1"
chat_with_history.invoke(
    {"user_question": "اسمي براء"},
    config={"configurable": {"session_id": "user1"}}
)
```

```python
import psycopg 
from langchain_postgres import PostgresChatMessageHistory

conn_str = "postgresql://postgres:qwe123@localhost:5432/test_cairo"  
connect_database = psycopg.connect(conn_str)

table_name = "chat_history_online_g1"
# run this only once to create the table in the database
PostgresChatMessageHistory.create_tables(connect_database, table_name)
```

```python
connect_database
```

```python
import uuid

user_id = str(uuid.uuid4())
user_id
```

```python

def get_history_from_postgres(user_id):
    return PostgresChatMessageHistory(
    table_name,
    user_id,
    sync_connection=connect_database
    )
```

```python
get_history_from_postgres(user_id)
```

```python
from langchain_core.runnables.history import RunnableWithMessageHistory

chat_with_history_postgres = RunnableWithMessageHistory(
    chat_chain,
    get_history_from_postgres, 
    input_message_key="user_question",
    history_message_key="History"
)
```

```python
question = "what is the capital of France?"
chat_with_history_postgres.invoke(
    {"user_question": "شو اسمي "},
    config={"configurable": {"session_id": user_id}}
)
```
