```python
import gradio as gr
import requests
import uuid

# API URL
url = "http://127.0.0.1:8005/chat_body"

# session ثابت لكل المستخدم
session_id = str(uuid.uuid4())

def chat_with_api(message, history):
    try:
        response = requests.post(
            url,
            json={
                "user_question": message,
                "user_id": session_id
            }
        )

        if response.status_code == 200:
            answer = response.json().get("response")["content"]
        else:
            answer = f"Error: {response.status_code}"
            

    except Exception as e:
        answer = f"Exception: {str(e)}"
        

    return answer

# واجهة المحادثة
demo = gr.ChatInterface(
    fn=chat_with_api,
    title="AI Chat System 🔥",
    description="Connected to FastAPI backend",
)

demo.launch()
```
